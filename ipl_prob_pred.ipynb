```python
import pandas as pd   # Importing libraries for data manipulation
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
match_df = pd.read_csv("matches.csv")   # Reading the matches dataset
```

```python
match_df.columns  # Checking columns
```

```python
delivery_df = pd.read_csv("deliveries.csv") # Reading the deliveries dataset
```

```python
delivery_df.columns  # Checking columns
```

```python
match_df.head() # Display 1st 5 rows of matches dataset
```

```python
delivery_df.head() # Display 1st 5 rows of deliveries dataset
```

```python
df = match_df.merge(delivery_df,left_on = 'id', right_on = 'match_id') # merging the deliveries and matches dataset
```

```python
df.head() # Display 1st 5 rows
```

```python
df.columns # Checking columns
```

```python
final_df = df[['id','inning', 'city', 'toss_winner', 'toss_decision', 'winner', 'target_runs', 'target_overs' , 'batting_team', 
       'bowling_team', 'over', 'ball', 'total_runs',
       'player_dismissed']]    # Selecting the required columns
```

```python
final_df.head() # Display 1st 5 rows
```

```python
final_df['player_dismissed'] = final_df['player_dismissed'].apply(lambda x : 0 if pd.isna(x) else 1)  # Convering the player_dismissed columns to 0 and 1
```

```python
final_df['player_dismissed'].unique() # Checking unique values 
```

```python
final_df.shape # checking the shape of the dataset
```

```python
final_df['current_run']=final_df.groupby(['inning','id'])['total_runs'].cumsum()  # calculating the current runs
```

```python
final_df['required_run'] = final_df['target_runs'] - final_df['current_run'] # calculating the required runs
```

```python
final_df['balls_left'] = 120 - final_df['over']*6 - final_df['ball']  # calculating the balls left
```

```python
final_df['player_down'] = final_df.groupby(['inning','id'])['player_dismissed'].cumsum() # calculating the wickets down
```

```python
final_df['wicket_left'] = 10 - final_df['player_down'] # calculating the wickets left 
```

```python
final_df['current_rr'] = (final_df['current_run']*6)/(120-final_df['balls_left'])  # calculating the current run rate
```

```python
final_df['required_rr'] = final_df['required_run']*6/final_df['balls_left'] # calculating the required runs rate
```

```python
final_df.head() # Display 1st 5 rows
```

```python
final_df.columns # Display columns
```

```python
final_df.shape # checking the shape of dataset
```

```python
final_df = final_df[(final_df['inning'].isin([1,2]))] # Filtering the dataset for only 1st and 2nd innings
```

```python
final_df.shape  # checking the shape of dataset
```

```python
inn_2 = final_df[final_df['inning'] == 2]  # Filtering the dataset for only 2nd innings
```

```python
inn_2.shape  # checking the shape of dataset
```

```python
inn_2.head() # Display 1st 5 rows
```

```python
def result(row):  # Function to determine if the batting team won or not
    return 1 if row['winner'] == row['batting_team'] else 0
```

```python
inn_2['result'] = inn_2.apply(result,axis = 1) # Applying the result function to the dataset
```

```python
inn_2.head()
```

```python
inn_2.columns # checking the columns
```

```python
final_data = inn_2[['batting_team','bowling_team','city','target_runs','required_run','balls_left','wicket_left','current_rr','required_rr','result']]
```

```python
final_data.head()
```

```python
final_data = final_data[~final_data['batting_team'].isin(['Kochi Tuskers Kerala','Pune Warriors','Gujarat Lions','Rising Pune Supergiants','Rising Pune Supergiant'])
& ~final_data['bowling_team'].isin(['Kochi Tuskers Kerala','Pune Warriors','Gujarat Lions','Rising Pune Supergiants','Rising Pune Supergiant'])]

                       # Removing the teams which are no longer the part of IPL
```

```python
final_data.shape
```

```python
final_data.replace(to_replace = 'Royal Challengers Bangalore', value = 'Royal Challengers Bengaluru',inplace = True)
final_data.replace(to_replace = 'Kings XI Punjab', value = 'Punjab Kings',inplace = True)
final_data.replace(to_replace = 'Delhi Daredevils', value = 'Delhi Capitals',inplace = True)
final_data.replace(to_replace = 'Deccan Chargers', value = 'Sunrisers Hyderabad',inplace = True)
     
# Replacing the old team names with new team names
```

```python
final_data = final_data.sample(final_data.shape[0]) # shuffling the dataset
```

```python
final_data.head() # Display 1st 5 rows
```

```python
final_data.isnull().sum() # checking for null values
```

```python
final_data.dropna(inplace=True) # removing null values
```

```python
final_data = final_data[final_data['balls_left']!=0] # removing the rows where balls left is 0 (it makes infinite values in required_rr)
```

```python
final_data.info() # checking the info of the dataset
```

```python
final_data.describe() # checking the statistical measures of the dataset
```

```python
final_data.shape # checking the shape of dataset
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score
 
             # Importing libraries for model building and evaluation
```

```python
x = final_data.drop(columns=['result']) # independent features
y = final_data['result'] # dependent features
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=5) # splitting the dataset into training and testing set
```

```python
trf = ColumnTransformer([  
    ('ohe',OneHotEncoder(sparse_output=False, drop='first'),['batting_team','bowling_team','city'])
],remainder='passthrough')  # Column transformer for one hot encoding the categorical features
```

```python
pipe = Pipeline(steps = [
    ('step1',trf),
    ('step2',LogisticRegression(solver = 'liblinear') )
])                                                        # Creating a pipeline for the model
```

```python
pipe.fit(x_train,y_train) # fitting the model
```

```python
y_pred = pipe.predict(x_test) # predicting the values
```

```python
accuracy_score(y_test,y_pred) # checking the accuracy of the model
```
